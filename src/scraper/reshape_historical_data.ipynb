```python
out_obj = {}
counties = list(weekly_2018.county)
for county in counties:
    out_obj[county] = {
        'ts': [],
        'apps_submitted_latest': 0,
        'ballots_accepted_latest': 0,
    }
    for week in range(1, 8):
        out_obj[county]['apps_submitted_latest'] += int(weekly_2018[weekly_2018['county'] == county]['week{}_requested'.format(week)])
        out_obj[county]['ballots_accepted_latest'] += int(weekly_2018[weekly_2018['county'] == county]['week{}_ab'.format(week)])
        out_obj[county]['ts'].append({
            'weeks_away': 7-week,
            'apps_submitted': out_obj[county]['apps_submitted_latest'],
            'ballots_accepted': out_obj[county]['ballots_accepted_latest'],
        })
        
    # TODO: Add election day
#     electionday_requested	electionday_ab
```
