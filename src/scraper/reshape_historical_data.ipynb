```python
out_obj = {'ts_statewide': []}
counties = list(weekly_2018.county)
```

```python
counties
```

```python
def process_week(ballots_requested, ballots_accepted, week_name):
    print(ballots_requested, ballots_accepted, week_name)
    out_obj[county]['apps_submitted_latest'] = ballots_requested
    out_obj[county]['ballots_accepted_latest'] = ballots_accepted

    ts_record = {
        'weeks_away': week_name,
        'apps_submitted': ballots_requested,
        'ballots_accepted': ballots_accepted,
    }

    out_obj[county]['ts'].append(ts_record.copy())

    ts_statewide_date = [date for date in out_obj['ts_statewide'] if date['weeks_away'] == week_name]
    if len(ts_statewide_date) == 0:
        out_obj['ts_statewide'].append(ts_record)
    else:
        # ts_statewide_date[0]['weeks_away'] += weeks_away(version_obj['data_date'])
        ts_statewide_date[0]['apps_submitted'] += out_obj[county]['apps_submitted_latest']
        ts_statewide_date[0]['ballots_accepted'] += out_obj[county]['ballots_accepted_latest']
```

```python
for county in counties:
    out_obj[county] = {
        'ts': [],
        'apps_submitted_latest': 0,
        'ballots_accepted_latest': 0,
    }
    for week in range(1, 8):
        ballots_requested = int(weekly_2018[weekly_2018['county'] == county]['week{}_requested'.format(week)])
        ballots_accepted = int(weekly_2018[weekly_2018['county'] == county]['week{}_ab'.format(week)])
        
        process_week(ballots_requested, ballots_accepted, 7-week)
        
    # Add election day
    ballots_requested = int(weekly_2018[weekly_2018['county'] == county]['electionday_requested'])
    ballots_accepted = int(weekly_2018[weekly_2018['county'] == county]['electionday_ab'])
    process_week(ballots_requested, ballots_accepted, 'election_day')
```
