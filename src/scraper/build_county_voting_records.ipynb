```python
import pandas as pd
```

```python
old_xls = pd.read_excel('import/mncounties.xlsx')
old_xls
```

```python
corrected_csv = pd.read_csv('import/mncounties.csv')
corrected_csv
```

```python
# Merge to take vote counts from csv, since we found some percentage errors in building the sliders
```

```python
merged_doc = corrected_csv.merge(
    old_xls[[
        'county',
        'population2018',
        'winner_2012',
        'winner_2016',
        'winner_2018'
    ]],
    how="left",
    on="county"
).drop(columns=['GEONAME'])
```

```python
merged_doc.loc[merged_doc['county'] == 'Wright', 'classification'] = 'rural'
```
